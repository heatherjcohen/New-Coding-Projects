```python
#Import ALL THE THINGS
import wikipedia 
import re 
import string 
from textblob import TextBlob
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
from sklearn import cross_validation
from collections import Counter
import gensim 
import random
```

```python
#Open files
a = open('fiveyears.txt', 'r')
b = open('hang on to yourself.txt', 'r')
c = open('it aint easy.txt', 'r')
d = open('lady stardust.txt', 'r')
e = open('moonage daydream.txt', 'r')
f = open('rock n roll.txt', 'r')
g = open('soul love.txt', 'r')
h = open('star.txt', 'r')
i = open('starman.txt', 'r')
j = open('suffragette city.txt', 'r')
k = open('ziggylyrics.txt', 'r')

#Read files
a_lyrics = a.read()
b_lyrics = b.read()
c_lyrics = c.read()
d_lyrics = d.read()
e_lyrics = e.read()
f_lyrics = f.read()
g_lyrics = g.read()
h_lyrics = h.read()
i_lyrics = i.read()
j_lyrics = j.read()
k_lyrics = k.read()
```

```python
lyrics = a_lyrics + b_lyrics + c_lyrics + d_lyrics + e_lyrics + f_lyrics + g_lyrics + h_lyrics + i_lyrics + j_lyrics + k_lyrics
```

```python
blob = TextBlob(lyrics)
```

```python
nplyrics = blob.words

#USE TO SHOW DENSITY OF NONSENSE NOISE
#Lexical Density? 
print float(len(blob.noun_phrases))/float(len(blob.words))
```

```python
#Change Input
nplyrics = blob.noun_phrases

vect = CountVectorizer(analyzer='char_wb', ngram_range=(1, 5), encoding=u'utf-8', stop_words='english', max_features=10000)
ngvText = vect.fit_transform(nplyrics)

from sklearn.decomposition import TruncatedSVD 

N_COMPONENTS =100 # feel free to change this if you so desire; typical values are between 50 and 400

tsvd = TruncatedSVD(n_components=N_COMPONENTS)

embedded_corpus = tsvd.fit_transform(ngvText)

from sklearn.cluster import MiniBatchKMeans 

N_CLUSTERS = 38  # feel free to experiment with this value, too

clstr = MiniBatchKMeans(n_clusters=N_CLUSTERS)
clusters = clstr.fit_predict(embedded_corpus)

from collections import defaultdict

clusters_to_content = defaultdict(list)

for subject, page in zip(nplyrics, clusters): 
    clusters_to_content[page].append(subject)

for idx in clusters_to_content: 
    print(idx)
    print(clusters_to_content[idx])
    print('\n')
    
```

```python
vect = CountVectorizer(analyzer='char_wb', ngram_range=(3, 4), encoding=u'utf-8', stop_words='english', max_features=10000)
ngvText = vect.fit_transform(nplyrics)

from sklearn.decomposition import TruncatedSVD 

N_COMPONENTS =100 # feel free to change this if you so desire; typical values are between 50 and 400

tsvd = TruncatedSVD(n_components=N_COMPONENTS)

embedded_corpus = tsvd.fit_transform(ngvText)

from sklearn.cluster import MiniBatchKMeans 

N_CLUSTERS = 25  # feel free to experiment with this value, too

clstr = MiniBatchKMeans(n_clusters=N_CLUSTERS)
clusters = clstr.fit_predict(embedded_corpus)

from collections import defaultdict

clusters_to_content = defaultdict(list)

for subject, page in zip(nplyrics, clusters): 
    clusters_to_content[page].append(subject)

for idx in clusters_to_content: 
    print(idx)
    print(clusters_to_content[idx])
    print('\n')
    
```
