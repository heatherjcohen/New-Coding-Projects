```python
#Import ALL THE THINGS
import wikipedia 
import re 
import string 
from textblob import TextBlob
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
from sklearn import cross_validation
from collections import Counter
import gensim 
```

```python
f = open('fiveyears.txt', 'r')
lyrics = f.read()
```

```python
vect = CountVectorizer(analyzer='char_wb', ngram_range=(3, 10), encoding=u'utf-8', stop_words='english', max_features=10000)
ngvText = vect.fit_transform(lyrics)
```

```python
words = re.findall(r'\w+', lyrics) #This finds words in the document

cap_words = [word.upper() for word in words] #capitalizes all the words

word_counts = Counter(cap_words)
```
